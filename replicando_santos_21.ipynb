```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings
from sklearn.preprocessing import MinMaxScaler
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('data/serie_cafe_cepea_mensal.xlsx')
df.rename(columns = {
    'Data': 'date',
    'À vista R$': 'price_brl',
    'À vista US$': 'price_usd'
}, inplace = True)
df["log_price_brl"] = np.log(df["price_brl"])
```

```python
# treatment implemented by santos

df_filtered = df.loc[
    (df['date'] >= '1997-01-01') & 
    (df['date'] <= '2021-12-01')
       ]
```

```python
scaler = MinMaxScaler()
price_scaled = scaler.fit_transform(df[['price_brl']])
```

```python
import numpy as np

def create_sequences(data, window=24):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i-window:i])
        y.append(data[i])
    return np.array(X), np.array(y)

X, y = create_sequences(price_scaled, window=24)
```

```python
X_train, X_test = X[:-6], X[-6:]
y_train, y_test = y[:-6], y[-6:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Bidirectional, LSTM, Dense

model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(24,1)),
    MaxPooling1D(pool_size=2),
    Bidirectional(LSTM(50)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```
